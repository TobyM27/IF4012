```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# load the img in bgr format
img = cv2.imread('tinyLuffy.jpg')

# convert img to bgr format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# display the image
cv2_imshow(img)
```

```python
# Convert BGR to RGB
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert RGB to grayscale
gray_img = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2GRAY)

# Convert grayscale to binary
ret, binary_img = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)

#showing the processed image using matplotlib plt
plt.subplot(1, 3, 1)
plt.imshow(rgb_img)
plt.subplot(1, 3, 2)
plt.imshow(gray_img)
plt.subplot(1, 3, 3)
plt.imshow(binary_img)
plt.show()

# Display the processed images using matplotlib
#cv2_imshow(rgb_img)
#cv2_imshow(gray_img)
#cv2_imshow(binary_img)
```

```python
# Convert img to BGR
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert img to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Convert img to binary
ret, binary_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

#showing the processed image using matplotlib plt
plt.subplot(1, 3, 1) # # from img to bgr_img
plt.imshow(rgb_img)
plt.subplot(1, 3, 2)
plt.imshow(gray_img) # from img to gray_img
plt.subplot(1, 3, 3)
plt.imshow(binary_img) # from img to binary_img
plt.show()
```

```python
# remove noise from greyscale
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(gray_img, cv2.MORPH_OPEN, kernel)
#cv2_imshow(opening)
#print("-----------------")

# apply adaptive threshold
adaptive_threshold = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# apply dilated image toward the adaptive_threshold
dilated_img = cv2.dilate(adaptive_threshold, kernel, iterations=1)

# apply contour detection toward dilated_img
contours, hierarchy = cv2.findContours(dilated_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# apply convex connected contour toward contour detection output
convex_hull = cv2.convexHull(contours[0])

# Create an empty image to draw the contour on
contour_img = np.zeros_like(img)
#cv2_imshow(contour_img)

# draw the convex connected contour on the empty image
cv2.drawContours(contour_img, [convex_hull], 0, (0, 255, 0), 2)

#show the output of convex_hull with using cv
#cv2_imshow(contour_img)

#show all the outputs with matplotlib with subplots
plt.figure(figsize=(15,5))
plt.subplot(1,5,1)
plt.imshow(opening)
plt.title('Opening')

plt.subplot(1,5,2)
plt.imshow(adaptive_threshold)
plt.title('Adaptive Threshold')

plt.subplot(1,5,3)
plt.imshow(dilated_img)
plt.title('Dilated Image')

plt.subplot(1,5,4)
plt.imshow(contour_img)
plt.title('Contour Image')

plt.subplot(1,5,5)
plt.imshow(img)
plt.title('Original Image')
plt.show()
```
